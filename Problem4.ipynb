A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
from itertools import combinations,permutations,product
def calc_number(x,y,z,C):
    return(C*100*x + C*10*y + C*z)

check_palin = lambda x: x == int(str(x)[::-1])
palins_lit = []


for C in range(100,1000):
    print(C)
    for x,y,z in product(range(10),repeat=3):
        result = calc_number(x,y,z,C)
        if check_palin(result):
            print(result)
            print("x: {}, y: {}, z: {}".format(x,y,z))
            if(x != 0):
                palins_lit.append(result)
    print("---------------------------------------")



       
```

```python
max(palins_lit)
```
